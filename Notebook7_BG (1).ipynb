11) Renombra la columna 'Life expectancy ' como 'life_expectancy', ' BMI ' como 'bmi' y ''Measles ' como 'measles'. No olvides de usar `inplace` para modificar el DataFrame original. Luego chequea que el cambio se haya hecho correctamente.

```python
# Renombro columnas del DataFrame que llame datos
datos.rename(columns={'Life expectancy ': 'life_expectancy', ' BMI ': 'bmi', 'Measles ': 'measles'}, inplace=True)

```

```python
# Chequeo que se hayan reemplazado correctamente
print(datos.columns)
```

```python
# Obtengo el promedio de la columna
prom = datos['life_expectancy'].mean().round(3) # Round redondea el valor
print(f"Promedio de la columna de life_expectancy es : {prom}")
```

```python
# Obtengo el promedio por pais
prom_pais = datos.groupby('Country')['life_expectancy'].mean()
# groupby va a agrupar los promedios en filas por pais

print(prom_pais)
```

```python
# Maxima expectativa de vida usando numpy
np.max(datos['life_expectancy'])
```

```python
# Minima expectativa de vida usando numpy
np.min(datos['life_expectancy'])
```

```python
# Minima expectativa de vida usando pandas
min = datos['life_expectancy'].min()
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al maximo

if max in datos['life_expectancy']:
  print(datos[datos['life_expectancy'] == max])
```
